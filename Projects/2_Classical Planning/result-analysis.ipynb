```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
import math
import numpy as np
```

```python
algos = ['BFS', 'DFS', 'UCS', 'GBFS h_unmet_goals', 'GBFS h_pg_levelsum', 'GBFS h_pg_maxlevel', 'GBFS h_pg_setlevel', 'A* h_unmet_goals', 'A* h_pg_levelsum', 'A* h_pg_maxlevel', 'A* h_pg_setlevel']
```

```python
nodes_bfs = [43, 343, 14663, 99736]
nodes_dfs = [21, 624, 408, 25174]
nodes_ucs = [60, 5154, 18510, 113339]
nodes_gbfs_unmet_goals = [7, 17, 25, 29]
nodes_gbfs_level_sum = [6, 9, 14, 17]
nodes_gbfs_max_level = [6, 27, 21, 56]
nodes_gbfs_set_level = [6, 9, 35, math.nan]
nodes_a_star_unmet_goals = [50, 2467, 7388, 34330]
nodes_a_star_level_sum = [28, 357, 369, 1208]
nodes_a_star_max_level = [43, 2887, 9580, math.nan]
nodes_a_star_set_level = [33, 1037, math.nan, math.nan]

# data to plot
n_problems = 4

# create plot
fig, ax = plt.subplots()
index = np.arange(n_problems)
bar_width = 0.08
opacity = 0.5

rects_bfs = plt.bar(index - 5 * bar_width, nodes_bfs, bar_width,
alpha=opacity,
color=(1,0,0),
label='BFS')

rects_dfs = plt.bar(index - 4 * bar_width, nodes_dfs, bar_width,
alpha=opacity,
color=(1,0,127/255),
label='DFS')

rects_ucs = plt.bar(index - 3 * bar_width, nodes_ucs, bar_width,
alpha=opacity,
color=(127/255,0,1),
label='UCS')

rects_gbfs_unmet = plt.bar(index - 2 * bar_width, nodes_gbfs_unmet_goals, bar_width,
alpha=opacity,
color=(0,0,1),
label='GBFS with h_unmet_goals')

rects_gbfs_sum = plt.bar(index - 1 * bar_width, nodes_gbfs_level_sum, bar_width,
alpha=opacity,
color=(0,127/255,1),
label='GBFS with h_pg_levelsum')

rects_gbfs_max = plt.bar(index + 0 * bar_width, nodes_gbfs_max_level, bar_width,
alpha=opacity,
color=(0,191/255,1),
label='GBFS with h_pg_maxlevel')

rects_gbfs_set = plt.bar(index + 1 * bar_width, nodes_gbfs_set_level, bar_width,
alpha=opacity,
color=(0,1,127/255),
label='GBFS with h_pg_setlevel')

rects_as_unmet = plt.bar(index + 2 * bar_width, nodes_a_star_unmet_goals, bar_width,
alpha=opacity,
color=(191/255,1,0),
label='A* with h_unmet_goals')

rects_as_sum = plt.bar(index + 3 * bar_width, nodes_a_star_level_sum, bar_width,
alpha=opacity,
color=(1,1,0),
label='A* with h_pg_levelsum')

rects_as_max = plt.bar(index + 4 * bar_width, nodes_a_star_max_level, bar_width,
alpha=opacity,
color=(1,191/255,0),
label='A* with h_pg_maxlevel')

rects_gbfs_set = plt.bar(index + 5 * bar_width, nodes_a_star_set_level, bar_width,
alpha=opacity,
color=(1,63/255,0),
label='A* with h_pg_setlevel')

plt.xlabel('Problem')
plt.ylabel('Nodes Expanded (log scale)')
plt.title('Complexity by problems and algorithms')
plt.xticks(index+bar_width/2, (['Problem 1', 'Problem 2', 'Problem 3', 'Problem 4']))
plt.xticks(rotation=90)
plt.yscale('log')
plt.legend()

plt.tight_layout()
#plt.show()
plt.savefig('repo-udacity-ai/Projects/2_Classical Planning/images/nodes_bar_chart.png')
```

```python
colors = {'BFS': (255/255,0,0), 'DFS': (255/255,0,127/255), 'UCS': (127/255,0,255/255),
          'GBFS h_unmet_goals': (0,0,255/255), 'GBFS h_pg_levelsum': (0,127/255,255/255), 'GBFS h_pg_maxlevel': (0,191/255,255/255), 'GBFS h_pg_setlevel': (0,255/255,17/255),
          'A* h_unmet_goals': (191/255,255/255,0), 'A* h_pg_levelsum': (255/255,255/255,0), 'A* h_pg_maxlevel': (255/255,191/255,0), 'A* h_pg_setlevel': (255/255,63/255,0)}
```

```python
df=pd.DataFrame({'actions': [20, 72, 88, 104], 'BFS': nodes_bfs, 'DFS': nodes_dfs, 'UCS': nodes_ucs,
                 'GBFS h_unmet_goals': nodes_gbfs_unmet_goals, 'GBFS h_pg_levelsum': nodes_gbfs_level_sum,
                 'GBFS h_pg_maxlevel': nodes_gbfs_max_level, 'GBFS h_pg_setlevel': nodes_gbfs_set_level,
                 'A* h_unmet_goals': nodes_a_star_unmet_goals, 'A* h_pg_levelsum': nodes_a_star_level_sum,
                 'A* h_pg_maxlevel': nodes_a_star_max_level, 'A* h_pg_setlevel': nodes_a_star_set_level,
                })
 
# multiple line plot
for algo in algos:
    plt.plot('actions', algo, data=df, marker='o', markerfacecolor=colors[algo], markersize=12, color=colors[algo], linewidth=4)

plt.xlabel('Actions in the problem')
plt.ylabel('Nodes Expanded (log scale)')
plt.title('Complexity by problems and algorithms')
plt.xticks(df.actions.unique())
plt.yscale('log')
plt.legend()

plt.tight_layout()
#plt.show()
plt.savefig('repo-udacity-ai/Projects/2_Classical Planning/images/nodes_lines.png')
```
